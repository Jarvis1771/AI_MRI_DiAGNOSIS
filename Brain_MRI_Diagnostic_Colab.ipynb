## 4️⃣ Launch the App 🚀

```
import subprocess
import time
from pyngrok import ngrok

# Start the FastAPI server in background
server = subprocess.Popen(
    ["python", "app.py"],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT
)

# Wait for server to start
print("⏳ Starting server...")
time.sleep(5)

# Create public tunnel
public_url = ngrok.connect(7860)

print("\n" + "=" * 60)
print("🧠 Brain MRI Diagnostic System is LIVE!")
print("=" * 60)
print(f"\n🌐 Public URL: {public_url}")
print(f"\n📋 Open the URL above in your browser to use the app.")
print(f"\n⚠️  Keep this cell running! Stopping it will shut down the app.")
print("=" * 60)

# Keep the cell alive and show server logs
try:
    while True:
        output = server.stdout.readline()
        if output:
            print(output.decode().strip())
        if server.poll() is not None:
            break
except KeyboardInterrupt:
    print("\n🛑 Server stopped.")
    server.terminate()
    ngrok.kill()
```

## 🛑 Stop Server (run this cell when done)

```
# Run this to stop the server and free resources
try:
    server.terminate()
    ngrok.kill()
    print("✅ Server stopped.")
except:
    print("Server already stopped.")
```
